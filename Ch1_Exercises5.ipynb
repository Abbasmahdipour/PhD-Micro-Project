3) Draw a two-dimensional diagram (COS)


```python
import numpy as np
x = np.linspace(0, 3*np.pi, 300)
y = np.cos(x)
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y, 'red')
```

```python
x = np.linspace(0, 3*np.pi, 50)
y = np.cos(x)
plt.plot(x, y, '-o')
```
